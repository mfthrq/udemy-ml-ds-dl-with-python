```python
import random
import numpy as np

def clusterData(n, k):
    random.seed(10)
    pointsPerCluster = float(n)/k 
    x = []
    for i in range(k):
        incomeCentroid = random.uniform(20000,200000)
        ageCentroid = random.uniform(80, 100)
        for j in range(int(pointsPerCluster)):
            x.append([np.random.normal(incomeCentroid, 10000), np.random.normal(ageCentroid, 2)])
    x = np.array(x)
    return x
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale 
```

```python
data = clusterData(100, 10)
model = KMeans(n_clusters=5)
model = model.fit(scale(data))
model.labels_
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(data[:,0], data[:,1], c=model.labels_.astype(np.float))
plt.show()
```
