```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm
```

Matplotlib Basics

```python
x = np.arange(-5, 5, 0.1)
plt.plot(x, norm.pdf(x))
plt.show()
```

Multiple plots one graph

```python
plt.plot(x, norm.pdf(x))
plt.plot(x, norm.pdf(x, 1, 0.5))
plt.show()
```

Adjust axes

```python
axes = plt.axes()
axes.set_xlim(-3,3)
axes.set_ylim(0,2)
plt.plot(x, norm.pdf(x))
plt.plot(x, norm.pdf(x, 1, 0.5))
plt.show()
```

Add grid

```python
axes = plt.axes()
axes.set_xlim(-3,3)
axes.set_ylim(0,2)
plt.plot(x, norm.pdf(x))
plt.plot(x, norm.pdf(x, 1, 0.5))
axes.grid()
plt.show()
```

Change Lines & Colors

```python
axes = plt.axes()
axes.set_xlim(-3,3)
axes.set_ylim(0,2)
plt.plot(x, norm.pdf(x), 'b--')
plt.plot(x, norm.pdf(x, 1, 0.5), 'r:')
plt.show()
```

Labeling axes and add legend

```python
axes = plt.axes()
axes.set_xlim(-3,3)
axes.set_ylim(0,2)
plt.plot(x, norm.pdf(x))
plt.plot(x, norm.pdf(x, 1, 0.5))
plt.grid()
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.legend(['This is Blue', 'This is orange'], loc=1)
# loc 1 = top right
# loc 2 = top left
# loc 3 = bottom left
# loc 4 = bottom right
plt.show()
```

Pie chart

```python
val = [20,20,40,20]
col = ['r', 'g', 'b', 'y']
exp = [0, 0, 0.02, 0]
labels = ['indo', 'india', 'singapore', 'germany']
plt.pie(val, colors=col, labels=labels, explode=exp)
plt.title("Negara")
plt.show()
```

Bar chart

```python
val = [30,25,40,55]
col = ['r', 'g', 'b', 'y']
plt.bar(range(0,4), val, color=col)
plt.show()
```

Scatter Plot

```python
x = np.random.randn(500)
y = np.random.randn(500)
plt.scatter(x,y)
plt.show()
```

Histogram

```python
gaji = np.random.normal(25000, 10000, 10000)
plt.hist(gaji, 20)
plt.show()
```

Box & Whisker Plot

```python
uniformSkewed = np.random.rand(100) * 100 - 40
high_outliers = np.random.rand(100) * 50 + 100
low_outliers = np.random.rand(100) * -50 + 100
data = np.concatenate((uniformSkewed, high_outliers, low_outliers))
plt.boxplot(data)
plt.grid()
plt.show()
```
