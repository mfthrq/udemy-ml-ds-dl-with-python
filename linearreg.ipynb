```python
import numpy as np
from matplotlib.pyplot import scatter
```

Linear relationship between page speed and purchase amounts

```python
pageSpeeds = np.random.normal(3.0, 1.0, 1000)
purchaseAmounts = 100 - (pageSpeeds + np.random.normal(0, 0.1, 1000)) * 3
scatter(pageSpeeds, purchaseAmounts)
```

```python
from scipy import stats

slope, intercept, r_value, p_value, std_err = stats.linregress(pageSpeeds, purchaseAmounts)
```

```python
r_value ** 2
```

use slope and intercept from the reg to plot predicted values vs observed

```python
import matplotlib.pyplot as plt
```

```python
def predict(x):
    return slope * x + intercept

fitline = predict(pageSpeeds)

plt.scatter(pageSpeeds, purchaseAmounts)
plt.plot(pageSpeeds, fitline, c='r')
plt.show()
```
